```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('images/exercicio_kp/obj02_001.jpg')
# img = cv2.imread('images/H07.png')
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_plt = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_plt)
```

```python
akaze = cv2.AKAZE_create(threshold=0.005)
kpts, desc = akaze.detectAndCompute(img, None)
```

```python
len(kpts)
```

```python
for marker in kpts:
    img_kp = cv2.drawMarker(img, tuple(int(i) for i in marker.pt), color=(0, 255, 0))
```

```python
img_kp = cv2.cvtColor(img_kp, cv2.COLOR_BGR2RGB)
```

```python
plt.rcParams['figure.figsize'] = [20, 30]
plt.imshow(img_kp)
```
