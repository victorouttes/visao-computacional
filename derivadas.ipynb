```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('images/cameraman.jpg')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img = cv2.GaussianBlur(gray,(3,3),0)
```

```python
sobx = cv2.Sobel(img, cv2.CV_64F, 1, 0)
```

```python
soby = cv2.Sobel(img, cv2.CV_64F, 0, 1)
```

```python
lap = cv2.Laplacian(img, cv2.CV_64F)
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
f, axarr = plt.subplots(2,2)
axarr[0,0].imshow(gray, cmap='gray')
axarr[0,1].imshow(sobx, cmap='gray')
axarr[1,0].imshow(soby, cmap='gray')
axarr[1,1].imshow(lap, cmap='gray')
```
